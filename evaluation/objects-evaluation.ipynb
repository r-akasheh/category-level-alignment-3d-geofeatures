```python
with open(base_path + "category-level-alignment-3d-geofeatures/evaluation/" + "ransac_mappings_objects_baseline_2.pickle", 'rb') as f:
    ransac_mappings_baseline = pickle.load(f)
```

```python
with open(base_path + "category-level-alignment-3d-geofeatures/evaluation/" + "ransac_mappings_objects_finetuned_2.pickle", 'rb') as f:
    ransac_mappings_finetuned = pickle.load(f)
```

```python
with open(base_path + "category-level-alignment-3d-geofeatures/evaluation/" + "teaser_mappings_objects_baseline.pickle", 'rb') as f:
    teaser_mappings_baseline = pickle.load(f)
```

```python
with open(base_path + "category-level-alignment-3d-geofeatures/evaluation/" + "teaser_mappings_objects_finetuned.pickle", 'rb') as f:
    teaser_mappings_finetuned = pickle.load(f)
```

```python
def turn_pickle_to_df(mappings):
    data = []
    # Iterate through each level of the dictionary
    for obj_name in mappings.keys():
        cos_theta = mappings[obj_name]['cos_theta']
        euclid_error = mappings[obj_name]['euclid_error']
        # Append a tuple with the collected data
        data.append((obj_name, cos_theta, euclid_error))

    # Create a DataFrame from the list
    df = pd.DataFrame(data,
                      columns=['obj_name', 'cos_theta', 'euclid_error'])
    return df
```

```python
turn_pickle_to_df(ransac_mappings_baseline)
```

```python
r_base = abs(turn_pickle_to_df(ransac_mappings_baseline).euclid_error).mean()
r_fine = abs(turn_pickle_to_df(ransac_mappings_finetuned).euclid_error).mean()
t_base = abs(turn_pickle_to_df(teaser_mappings_baseline).cos_theta).mean()
t_fine = abs(turn_pickle_to_df(teaser_mappings_finetuned).cos_theta).mean()
```
